**Лабораторная работа по языку Python**

Написать программу на языке Python, которая будет представлять собой простейшую СУБД.

Программа должна иметь текстовое меню, позволяющее пользователю выбирать операцию, которую он хочет совершить. Среди операций должны быть:
1.	Загрузка базы из файла формата csv.*
2.	Сохранение базы в файл формата csv.*
3.	Вывод содержимого базы на экран
4.	Добавление новой записи (строки) в базу
5.	Поиск и вывод на экран записи по ключевому полю
6.	Удаление записи (строки) по ключевому полю
7.	Выполнение индивидуального задания

База должна храниться в программе в виде списка объектов.


```python
import pandas as pd
import numpy as np
```

```python
def read_file(path):
  df = pd.read_csv(path, sep = ",")
  return df
```

```python
def save_file(df):
  try:
    path = input("Введите адрес сохранения: ")
    df.to_csv(path)
  except:
    print("Ошибка, нет данных!")
```

```python
def show_file(df):
  try:
    print(df.to_string())
  except:
    print("Ошибка, нет данных!")
```

```python
def add_value(df):
  try:
    init = input("Enter init: ")
    group = input("Enter group: ")
    mark = int(input("Enter mark: "))
    df.loc[len(df)] = [init, group, mark]
  except:
    print("Ошибка, нет данных!")
```

```python
def find(df):
  try:
    keyword = input("Введите ФИО: ")
    selected_row = df[df['init'] == keyword]
    print(selected_row)
  except:
    print("Ошибка, нет данных!")
```

```python
def delete_str(df):
  try:
    keyword = input("Введите ФИО: ")
    index = df.index[df['init'] == keyword]
    df.drop(index, inplace = True)
  except:
    print("Ошибка, нет данных!")
```

```python
def aggregate_by(df):
  unco = df.groupby(df.columns[1]).size()
  print(unco)
```

```python
df = pd.read_csv("class.csv", sep = ',')
unco = df['group'].count()
print(unco)
```

```python
def ind_task(df):
  marks = df['mark'].unique()
  count = df['group'].count()

  def quick_sort(marks):
      if len(marks)<=1: return marks
      else:
        pivot = marks[0]
        left_side = [i for i in marks[1:] if i<=pivot]
        right_side = [i for i in marks[1:] if i>pivot]
        return quick_sort(left_side) + [pivot] + quick_sort(right_side)
  Ls = quick_sort(marks)
  print(str(df.columns[1]) + " - " + str(count) + " - " + str(Ls[0]) + " - " + str(Ls[-1]))
```

```python
path = input("Введите адрес таблицы: ")
df = pd.DataFrame()
while True:
  print("\nМеню:")
  print("1 - Загрузить базу из файла")
  print("2 - Сохранить базу в файл")
  print("3 - Вывести содержимое")
  print("4 - Добавить учащегося")
  print("5 - Найти по ФИО")
  print("6 - Удалить по ФИОУ")
  print("7 - Агрегировать по группе")
  print("8 - Вывести статистику по 2 и 3 столбцам")
  print("0 - Выход")

  try:
    cmd = int(input("Введите номер команды: "))
  except ValueError:
    print("Введите корректный номер.")
    continue

  if cmd == 1:
    df = read_file(path)
  elif cmd == 2:
    save_file(df)
  elif cmd == 3:
    show_file(df)
  elif cmd == 4:
    add_value(df)
  elif cmd == 5:
    find(df)
  elif cmd == 6:
    delete_str(df)
  elif cmd == 7:
    aggregate_by(df)
  elif cmd == 8:
    ind_task(df)
  elif cmd == 0:
    print("Выход из программы.")
    break
  else:
    print("Неизвестная команда")
```
